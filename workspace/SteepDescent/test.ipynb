### 最急降下法について学ぶ

最急降下法とは、最適化（関数の最小化または最大化）の有名な手法の一つ。

#### 最急降下法の概要

1. 適当に初期点を選ぶ
2. 今いる位置における最急降下方向（関数の値が最も小さくなる方向）を計算する
3. 最急降下方向に進む
4. 2と3を繰り返す

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline

sns.set()

x = np.linspace(0, 4, 40)
y = x**2 - 4 * x + 5

plt.plot(x, y);
```
